# Main code for profanity testing

```python
from datasets import Dataset
dataset = Dataset.from_pandas(final_df)
dataset = dataset.train_test_split(test_size=0.2)
train_ds = dataset["train"]
test_ds = dataset["test"]
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
def tokenize_fn(example):
    inputs = tokenizer(
        example["textp"],
        padding="max_length",
        truncation=True,
        max_length=128
    )

    targets = tokenizer(
        example["label"],
        padding="max_length",
        truncation=True,
        max_length=5
    )

    inputs["labels"] = targets["input_ids"]
    return inputs
```

```python
train_ds = train_ds.map(tokenize_fn, batched=True, remove_columns=train_ds.column_names)
test_ds = test_ds.map(tokenize_fn, batched=True, remove_columns=test_ds.column_names)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./flan-t5-profanity",
    eval_strategy="epoch",
    logging_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=5e-5,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    save_total_limit=2,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    tokenizer=tokenizer
)

trainer.train()
```

```python
model.save_pretrained("./fine-tuned-flan-t5-profanity")
tokenizer.save_pretrained("./fine-tuned-flan-t5-profanity")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_path = "./fine-tuned-flan-t5-profanity"
tokenizer = T5Tokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)

strr = "What the hell do you want? I've told you assholes to leave me alone!"
sample_input = f"Classify if the message contains profanity: ${strr}"

inputs = tokenizer(
    sample_input,
    return_tensors="pt",
    padding="max_length",
    truncation=True,
    max_length=128
)

output_ids = model.generate(
    input_ids=inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_length=5
)

prediction = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print(f"Prediction: {prediction}")
```
